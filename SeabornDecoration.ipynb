```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

```python
def randplot(flip=1):
    for i in range(1, 7):
        plt.plot(np.random.randn(50).cumsum());
```

```python
randplot();
```

```python
sns.set()
randplot();
```

## Remove Axis Spin

```python
d = np.random.randn(50).reshape(10 ,5)
d
```

```python
sns.distplot(d)
sns.despine();
```

```python
sns.violinplot(data=d)
sns.despine(offset=10, trim=True);
```

```python
sns.boxplot(data=d, palette='deep')
sns.despine(left=True);
```

## Set Design Temporarily

```python
f = plt.figure(figsize=(6, 6))
gs = f.add_gridspec(2, 2)

with sns.axes_style('darkgrid'):
    ax = f.add_subplot(gs[0, 0])
    randplot()
    
with sns.axes_style('white'):
    ax = f.add_subplot(gs[0, 1])
    randplot()
    
with sns.axes_style('ticks'):
    ax = f.add_subplot(gs[1, 0])
    randplot()
    
with sns.axes_style('whitegrid'):
    ax = f.add_subplot(gs[1, 1])
    randplot()
    
f.tight_layout();
```

xkcd Color Table: https://xkcd.com/color/rgb/

```python
plt.plot(np.random.randn(50).cumsum(), sns.xkcd_rgb['grey'], lw=3)
plt.plot(np.random.randn(50).cumsum(), sns.xkcd_rgb['red'], lw=3)
plt.plot(np.random.randn(50).cumsum(), sns.xkcd_rgb['orange'], lw=3)
plt.plot(np.random.randn(50).cumsum(), sns.xkcd_rgb['pink'], lw=3)
```

```python
colors = ['soft blue', 'cerise', 'light olive green', 'pale orange', 'royal purple', 'teal']
sns.palplot(sns.xkcd_palette(colors))
```

```python
sns.palplot(sns.color_palette('RdYlGn'))
sns.palplot(sns.color_palette('nipy_spectral'))
sns.palplot(sns.color_palette('Accent'))
```

```python
sns.palplot(sns.color_palette('cubehelix', 8))
sns.palplot(sns.cubehelix_palette(8))
sns.palplot(sns.cubehelix_palette(8, start=.5, rot=.75))
sns.palplot(sns.cubehelix_palette(8, start=.5, rot=0, dark=0, light=.95, reverse=True))
```

```python
x, y = np.random.multivariate_normal([0, 0], [[1, -.5], [-.5, 1]], size=500).T
cmap = sns.cubehelix_palette(light=1, as_cmap=True)
sns.kdeplot(x, y, cmap=cmap, shade=True);
```

```python
sns.palplot(sns.light_palette('green'))
sns.palplot(sns.dark_palette('purple'))
sns.palplot(sns.light_palette('blue', reverse=True))
sns.palplot(sns.light_palette((210, 90, 60), input='husl'))
sns.palplot(sns.light_palette('muted purple', input='xkcd'))
```

```python
pal = sns.dark_palette('palegreen', as_cmap=True)
sns.kdeplot(x, y, cmap=pal);
```

```python
sns.palplot(sns.color_palette('RdYlGn', 7))
sns.palplot(sns.color_palette('BuPu', 7))
sns.palplot(sns.color_palette('YlOrBr', 7))
```
